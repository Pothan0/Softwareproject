```python
# Cell 2: Import libraries
import os
import textwrap
import chromadb
import requests
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
# Cell 3: Load PDF files
def load_pdf_files(file_path):
    """Loads a single PDF file."""
    # Use PyPDFLoader directly to load a single PDF file.
    loader = PyPDFLoader(file_path)
    documents = loader.load()
    return documents

data_directory = "/content/Medical_Book.pdf"  # Adjust if needed
extracted_data = load_pdf_files(data_directory)
```

```python
# Cell 4: Split text into chunks
def split_text(documents):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
    text_chunks = text_splitter.split_documents(documents)
    return text_chunks

text_chunks = split_text(extracted_data)
```

```python
# Cell 5: Initialize SentenceTransformer model
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Cell 6: Initialize ChromaDB
chroma_client = chromadb.PersistentClient(path="chroma_db")
collection = chroma_client.get_or_create_collection(name="medical_docs")
```

```python
# Cell 7: Store text chunks in ChromaDB
def store_in_chromadb(text_chunks):
    for idx, chunk in enumerate(text_chunks):
        text = chunk.page_content
        vector = embedding_model.encode(text).tolist()

        collection.add(
            ids=[str(idx)],
            documents=[text],
            embeddings=[vector]
        )
    print(f"Stored {len(text_chunks)} documents in ChromaDB!")

store_in_chromadb(text_chunks)
```

```python
# Cell 8: Convert text chunks to conversation data
import json
import random

def convert_to_conversation(text_chunks):
    conversation_templates = [
        ("Patient: Doctor, I have been experiencing {symptom}. What could be the reason?",
         "Doctor: {explanation} {treatment}")
    ]

    conversations = []

    for chunk in text_chunks:
        text = chunk.page_content
        sentences = text.split(". ")
        if len(sentences) < 2:
            continue

        symptom = sentences[0].strip()
        explanation = sentences[1].strip()
        treatment = sentences[2].strip() if len(sentences) > 2 else "You should consult a healthcare professional for proper treatment."

        template = random.choice(conversation_templates)
        patient_statement = template[0].format(symptom=symptom)
        doctor_response = template[1].format(explanation=explanation, treatment=treatment)

        conversations.append({"text": f"{patient_statement}\n{doctor_response}"})

    return conversations

conversation_data = convert_to_conversation(text_chunks)

output_file = "/content/medical_conversations.json" #adjust path if needed.
with open(output_file, "w") as f:
    json.dump(conversation_data, f, indent=4)

print(f"Converted {len(conversation_data)} medical text chunks into conversations and saved to {output_file}!")
```

```python
# Cell: Install dependencies (run first in Colab)
!pip install -q transformers datasets peft accelerate bitsandbytes

# Cell: Import libraries
import torch
import json
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer, BitsAndBytesConfig
from datasets import Dataset
from peft import LoraConfig, get_peft_model

# Cell: Load and process your JSON data
with open("/content/medical_conversations.json", "r") as f:
    raw_data = json.load(f)

# Convert to list of training examples
conversation_data = []
for entry in raw_data:
    convo = entry["conversations"]
    prompt = ""
    for turn in convo:
        role = "User" if turn["from"] == "human" else "Assistant"
        prompt += f"{role}: {turn['value']}\n"
    conversation_data.append({"text": prompt.strip()})

# Cell: Tokenizer & Model setup
model_name = "mistralai/Mistral-7B-Instruct-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map="auto"
)

# Cell: LoRA setup
lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj", "gate_proj", "down_proj", "up_proj"]
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# Cell: Prepare dataset
def prepare_dataset(conversations, tokenizer):
    dataset = []
    for conversation in conversations:
        text = conversation["text"]
        tokenized_text = tokenizer(text, truncation=True, padding="max_length", max_length=512)
        dataset.append(tokenized_text)
    return Dataset.from_list(dataset)

dataset = prepare_dataset(conversation_data, tokenizer)

# Cell: TrainingArguments
training_args = TrainingArguments(
    output_dir="./medical_mistral_finetuned",
    overwrite_output_dir=True,
    num_train_epochs=1,  # Start with 1 for testing
    per_device_train_batch_size=1,  # Keep low for Colab
    gradient_accumulation_steps=2,
    save_steps=50,
    save_total_limit=2,
    learning_rate=2e-4,
    logging_dir="./logs",
    logging_steps=5,
    fp16=True,
    optim="paged_adamw_32bit",
    lr_scheduler_type="cosine",
    warmup_ratio=0.03,
    group_by_length=True,
    push_to_hub=False,
)

# Cell: Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    tokenizer=tokenizer,
)

# ✅ Ready to call `trainer.train()` in the next cell
```

```python
# Cell 11a: Model Loading and Setup
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel

# Load base model and tokenizer
model_name = "mistralai/Mistral-7B-Instruct-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# Load model in 4-bit
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    llm_int8_enable_fp32_cpu_offload=True,
)

base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map="auto"
)

# Load LoRA adapters
model = PeftModel.from_pretrained(base_model, "./medical_mistral_lora_adapters")
model.eval()
```

```python
# Cell 11b: Response Generation Function (Modified)
import torch
import textwrap

def generate_response_with_finetuned_mistral(user_query, retrieved_text):
    prompt = f"""
    You are a medical assistant. Answer based on the retrieved text below:

    {retrieved_text}

    User Query: {user_query}
    """
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=500,

        )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = response.replace(prompt, "").strip() # remove prompt and strip extra spaces.
    response = response.replace("\n", " ").strip() # remove new lines.
    return response

def chatbot_response(user_query):
    if "hello" in user_query.lower():
        return "Hello! How can I assist you today?" # more natural greeting.

    query_vector = embedding_model.encode(user_query).tolist()
    results = collection.query(query_embeddings=[query_vector], n_results=1)
    retrieved_text = "\n".join(results["documents"][0]) if results["documents"] else "Thinking..."
    final_response = generate_response_with_finetuned_mistral(user_query, retrieved_text)
    return textwrap.fill(final_response, width=80)
```

```python
# Cell 11c: Chat Loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit", "bye", "thanks", "thank you"]:
        print("Goodbye!")
        break
    response = chatbot_response(user_input)
    print("\nMedibot:", response)
```
